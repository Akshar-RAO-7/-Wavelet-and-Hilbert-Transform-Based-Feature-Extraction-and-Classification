```python
import pywt

def wavelet_transform_row(signal_row, wavelet='db4', level_=1):

  """
  Applies Discrete Wavelet Transform (DWT) to a single row of data
  with user-specified level.

  Args:
      signal_row (pandas.Series or list): A single row of signal data.
      wavelet (str, optional): The wavelet function to use. Defaults to 'db4'.
      level (int, optional): The level of decomposition for DWT. Defaults to 1.

  Returns:
      tuple: A tuple containing the approximation coefficients (cA)
          and detail coefficients (cD) at different scales.
  """

  # Apply DWT with user-specified level
  wavelet_coeffs = pywt.wavedec(signal_row, wavelet, level=level_)

  # reutrn ac, dc
  return wavelet_coeffs[0], wavelet_coeffs[1:]


```

```python
import numpy as np

def hilbert_transform_row(signal_row):
    """
    Applies Hilbert Transform (HWT) to a single row of data.

    Args:
        signal_row (pandas.Series or list): A single row of signal data.

    Returns:
        numpy.ndarray: The transformed signal (real part of the Hilbert transform).
    """

    # Ensure signal is a NumPy array
    signal_array = np.asarray(signal_row)

    # Calculate DFT and zero-pad for imaginary component
    fft = np.fft.fft(signal_array)
    fft_len = len(fft)
    imag_component = np.zeros(fft_len, dtype=complex)

    # Set imaginary component for positive frequencies
    for i in range(1, int(fft_len / 2) + 1):
        imag_component[i] = 1j * fft[i]
    imag_component[0] = 0  # Imaginary part of DC component is 0

    # Set imaginary component for negative frequencies (mirrored)
    for i in range(1, int(fft_len / 2)):
        index = fft_len - i - 1
        imag_component[index] = -1j * fft[i]
    imag_component[fft_len // 2] = 0j  # Handle Nyquist frequency

    # Combine real and imaginary components
    hilbert_transform = fft + imag_component

    # Apply inverse DFT
    hilbert_transform = np.fft.ifft(hilbert_transform)

    # Return the real part of the Hilbert transform
    return hilbert_transform.real

# Example usage (replace data_row with your actual data row)
transformed_data = hilbert_transform_row([0,10,22,3,4,5])
print(transformed_data)
```
